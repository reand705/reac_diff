```python
# -*- coding: utf-8 -*-
# <nbformat>3.0</nbformat>

# <codecell>

# coding: utf-8
%matplotlib inline
import matplotlib
import scipy.integrate
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import rc ## desnecessário
matplotlib.rcParams['text.usetex'] = True
matplotlib.rcParams.update({'font.size': 18})

plt.figure(figsize=(10,8))
```

```python
#Imprima os diagramas de bifurcação que compara a integração númerica direta da EDP com as EDOs do método variacional
#Atenção para os nomes dos arquivos

#Presa

plt.figure(figsize=(10,8))

DP = [0.01,0.05,0.1]
DN = [0.0 + 0.001*i for i in range(0,201)]

#Variacional
for i in range(len(DP)):
    maxA1 = []
    minA1 = []
    for j in range(len(DN)):
        sol = np.load('Var_Method_Ros_Mac_dp=%.3f,dn=%.3f.npy' % (DP[i],DN[j]))
        A1 = sol[:,0]
        maxA1.append(np.max(A1[-1000:]))
        minA1.append(np.min(A1[-1000:]))
    plt.plot(DN,maxA1, color = (DP[i], DP[i]*10, DP[i]*10))
    plt.plot(DN,minA1, color = (DP[i], DP[i]*10, DP[i]*10))
    
DN = [0.0 + 0.002*i for i in range(0,101)]

plt.axis([0.0,0.20,0,6])
plt.title(r"$n$")
plt.xlabel(r"$d_n$")
plt.ylabel(r"$n_{1/2}$")

#Integração numérica
M = 101
for i in range(len(DP)):
    maxu = []
    minu = []
    for j in range(len(DN)):
        sol = np.load('Method_Of_Lines_dp=%.3f,dn=%.3f.npy' % (DP[i],DN[j]))
        u = np.transpose(sol[:,:M])
        maxu.append(np.max(u[M/2,-1000:]))
        minu.append(np.min(u[M/2,-1000:]))
    plt.plot(DN,maxu, marker = '.', linestyle="None", color = (DP[i], DP[i]*10, DP[i]*10), label = DP[i])
    plt.plot(DN,minu, marker = '.', linestyle="None", color = (DP[i], DP[i]*10, DP[i]*10))
    

plt.legend(loc='best', prop={'size': 20})
```

```python
#Predador

DP = [0.01,0.05,0.1]
DN = [0.0 + 0.001*i for i in range(0,201)]

plt.figure(figsize=(10,8))

#Variacional
for i in range(len(DP)):
    maxA2 = []
    minA2 = []
    for j in range(len(DN)):
        sol = np.load('Var_Method_Ros_Mac_dp=%.3f,dn=%.3f.npy' % (DP[i],DN[j]))
        A2 = sol[:,1]
        maxA2.append(np.max(A2[-1000:]))
        minA2.append(np.min(A2[-1000:]))
    plt.plot(DN,maxA2, color = (DP[i], DP[i], DP[i]*10))
    plt.plot(DN,minA2, color = (DP[i], DP[i], DP[i]*10))
    
DN = [0.0 + 0.002*i for i in range(0,101)]

plt.axis([0.0,0.20,0,11])
plt.title(r"$p$")
plt.xlabel(r"$d_n$")
plt.ylabel(r"$p_{1/2}$")

#Integração numérica
M = 101
for i in range(len(DP)):
    maxv = []
    minv = []
    for j in range(len(DN)):
        sol = np.load('Method_Of_Lines_dp=%.3f,dn=%.3f.npy' % (DP[i],DN[j]))
        v = np.transpose(sol[:,M:])
        maxv.append(np.max(v[M/2,-1000:]))
        minv.append(np.min(v[M/2,-1000:]))
    plt.plot(DN,maxv, marker = '.', linestyle="None", color = (DP[i], DP[i], DP[i]*10), label = DP[i])
    plt.plot(DN,minv, marker = '.', linestyle="None", color = (DP[i], DP[i], DP[i]*10))

plt.legend(loc='best', prop={'size': 20})
```
